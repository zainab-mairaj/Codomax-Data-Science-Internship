# Day 7: Data Analysis & Business

```python
import pandas as pd

# load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset loaded successfully! Total records:", len(df))
```

```python
# Calculating key metrics
total_passengers = df['PassengerId'].count()
total_fare_collected = df['Fare'].sum()
average_fare = df['Fare'].mean()
min_fare = df['Fare'].min()
max_fare = df['Fare'].max()

average_age = df['Age'].mean()

# Displaying Business Insights
print("=== 📊 TITANIC BUSINESS INSIGHTS SUMMARY ===")
print(f"Total Passengers Count: {total_passengers}")
print(f"Total Revenue Generated (Fare): ${total_fare_collected:.2f}")
print(f"Average Ticket Price: ${average_fare:.2f}")
print(f"Minimum Ticket Price: ${min_fare:.2f}")
print(f"Maximum Ticket Price: ${max_fare:.2f}")
print(f"Average Passenger Age: {average_age:.1f} years")
```
