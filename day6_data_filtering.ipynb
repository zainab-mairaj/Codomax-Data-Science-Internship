# day 6: data filtering, selection

```python
import pandas as pd

# Dataset load karna
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset loaded successfully! Total rows:", len(df))
```

# task 1: filtering rows (conditions)

```python
# Filtering: Age > 30 AND Pclass == 1
filtered_rows = df[(df['Age'] > 30) & (df['Pclass'] == 1)]

print("--- Filtered Passengers (Age > 30 & Class 1) ---")
print(filtered_rows.head())
print("\nFiltered Rows Count:", len(filtered_rows))
```

# task 2: selecting specific columns 

```python
# Selecting specific columns: Name, Sex, Age, Fare
selected_data = filtered_rows[['Name', 'Sex', 'Age', 'Fare']]

print("--- Selected Columns Only ---")
print(selected_data.head())
```

# task 4: sorting the dataset

```python
# Sorting by 'Fare' in descending order (highest to lowest)
sorted_dataset = selected_data.sort_values(by='Fare', ascending=False)

print("--- Final Filtered & Sorted Dataset (Highest Fare First) ---")
print(sorted_dataset.head())
```
