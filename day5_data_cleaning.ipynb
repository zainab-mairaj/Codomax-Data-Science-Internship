# Day 5: Data Cleaning & Preprocessing

```python
import pandas as pd

#Load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Initial Dataset Shape (Rows, Columns):", df.shape)
```

# Task 1: Handle Missing Values

```python
# check total missin values per column
print("--- Missing Values Before Cleaning ---")
print(df.isnull().sum())

# 1. fill missing 'Age' values with the median age
df['Age'] = df['Age'].fillna(df['Age'].median())

# 2. Drop 'Cabin' column (too many missing values to accurately fill)
df = df.drop(columns=['Cabin'])

# 3. Drop remianing rows with missing values (e.g., in 'Embarked')
df = df.dropna()

print("\n--- Missing Values After Cleaning ---")
print(df.isnull().sum())
```

# Task 2: Remove Duplicate Records

```python
#check number of duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# remove duplicate rows (if any exist)
df = df.drop_duplicates()

print("Dataset shape after removing duplicates:", df.shape)
```

# Task 3: Correct Data Types

```python
# Check initial data types
print("--- Data Types Before ---")
print(df[['PassengerId', 'Survived', 'Age', 'Fare']].dtypes)

# Convert 'Age' from float to integer
df['Age'] = df['Age'].astype(int)

# Convert 'Survived' to categorical type
df['Survived'] = df['Survived'].astype('category')

print("\n--- Data Types After Correction ---")
print(df[['PassengerId', 'Survived', 'Age', 'Fare']].dtypes)
print("\nClean Dataset Prepared Successfully!")
```
