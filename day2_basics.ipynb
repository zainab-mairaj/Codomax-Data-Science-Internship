# Challenge 1: Productivity Score 

```python
datasets = int(input("Enter datasets cleaned today: "))
hours = float(input("Enter hours worked: "))

efficiency = datasets / hours

print(f"Productivity Score: {efficiency:.2f} datasets/hours")
print(f"Types - Datasets: {type(datasets)}, Hours: {type(hours)}")
```

# Challenge 2: The Logic Builder

```python
col_name = str(input("Enter your column name: "))

if "#" in col_name or "$" in col_name:
    print("Data Status: Error: Invalid Characters")
elif col_name.islower():
    print("Data Status: Standard format")
else:
    print("Data Status: Needs Review")
```
