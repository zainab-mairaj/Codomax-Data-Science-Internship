# Day 8: DATA VISUALIZATION USING MATPLOTLIB

```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Libraries imported and dataset loaded successfully!")
```

# task 1: bar chart (passenger count by class)

```python

class_counts = df['Pclass'].value_counts()


plt.figure(figsize=(7, 4))
plt.bar(class_counts.index.astype(str), class_counts.values, color=['#1f77b4', '#ff7f0e', '#2ca02c'])

plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# task 2: pie chart (survival rate percentage)

```python

survived_counts = df['Survived'].value_counts()


plt.figure(figsize=(6, 6))
plt.pie(
    survived_counts.values, 
    labels=['Died (0)', 'Survived (1)'], 
    autopct='%1.1f%%', 
    colors=['#e74c3c', '#2ecc71'], 
    startangle=90
)

plt.title('Titanic Survival Percentage')
plt.show()
```

# task 3: line chart (ticket fares trend)

```python

sample_df = df.head(50)


plt.figure(figsize=(9, 4))
plt.plot(sample_df.index, sample_df['Fare'], color='purple', marker='o', linewidth=1.5, markersize=4)

plt.title('Ticket Fare Trend (First 50 Passengers)')
plt.xlabel('Passenger Index')
plt.ylabel('Fare Paid ($)')
plt.grid(True, linestyle=':', alpha=0.6)

plt.show()
```
